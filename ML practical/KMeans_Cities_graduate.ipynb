```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("C:\\Users\\muggs\\Desktop\\ML Dataset\\LP-1 Dataset\\K-Means & Hirarchical Clustering\\cities_r2.csv")
data.head(10)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
selected_col = data[['total_graduates']]
```

```python
scaler = StandardScaler()
selected_col = scaler.fit_transform(selected_col)
```

```python
cluster = []
for i in range(1,25):
    kmeans = KMeans(n_clusters=i,n_init=10,random_state=10)
    kmeans.fit(selected_col)
    cluster.append(kmeans.inertia_)
```

```python
cluster
```

```python
plt.plot(range(1,10),cluster,color = 'blue',marker = 'o', markerfacecolor ='red')
plt.xlabel('No of cluster')
plt.ylabel('cluster score')
plt.show()
```

```python
kmeans=KMeans(3,random_state=10)
kmeans.fit(selected_col)
```

```python
pred = kmeans.predict(selected_col)
pred
```

```python
data['clusters']=pred
```

```python
data.head(10)
```

```python
plt.scatter(selected_col,data['population_total'],c=data['clusters'],marker='o')
plt.xlabel('total grads')
    
```
